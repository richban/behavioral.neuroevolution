```python
from robot.vrep_robot import VrepRobot
from vrep.handlers import set_pose, get_object_handle
import robot.vrep as vrep
import numpy as np
import time
```

```python
robot_data = np.load('thymio_data.dat')
```

```python
OP_MODE = vrep.simx_opmode_oneshot_wait
```

```python
vrep.simxFinish(-1)
CLIENT_ID = vrep.simxStart(
    '127.0.0.1',
    19997,
    True,
    True,
    5000,
    5)
```

```python
robot = VrepRobot(CLIENT_ID, None, OP_MODE)
```

```python
vrep.simxStartSimulation(CLIENT_ID, vrep.simx_opmode_oneshot)
```

```python
robot.get_position()
```

```python
dummy = [0.94552579522132874, 0.4515371441841125, 0.93866877555847168]
```

```python
robot_data
```

```python
r = get_object_handle(CLIENT_ID, 'Pioneer_p3dx')
```

```python
i = 0
while i < len(robot_data):
    set_pose(CLIENT_ID, r, robot_data[i]['position'], vrep.simx_opmode_oneshot_wait)
```

```python
r, angles = vrep.simxGetObjectOrientation(CLIENT_ID, r, -1, vrep.simx_opmode_oneshot_wait)
```

```python
vrep.simxStopSimulation(CLIENT_ID, OP_MODE)
```
